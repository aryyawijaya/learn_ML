## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Import Dataset

### Import winequality-red.csv

```python
df_red = pd.read_csv('winequality-red.csv', sep=';')
df_red.head()
```

#### Add Column wine_type

```python
df_red['wine_type'] = 'Red Wine'
df_red.head()
```

### Import winequality-white.csv

```python
df_white = pd.read_csv('winequality-white.csv', sep=';')
df_white.head()
```

#### Add Column wine_type

```python
df_white['wine_type'] = 'White Wine'
df_white.head()
```

## Concatenate df_red and df_white

```python
df = pd.concat([df_red, df_white], ignore_index=True)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['quality'].unique()
```

## Data Preprocessing

### Grouping Values of Quality to Low (3-4), Medium (5-7), High (8-9)

```python
df['quality'] = df['quality'].map({3: 'low', 4: 'low', 5: 'medium', 6: 'medium', 7: 'medium', 8: 'high', 9: 'high'})
df.head()
```

### Encode Group of Quality to 0 (Low), 1 (Medium), 2 (High)

```python
df['quality'] = df['quality'].map({'low': 0, 'medium': 1, 'high': 2})
df.head()
```

## Data Visualization

```python
plt.figure(figsize=(11, 7))
sns.heatmap(df.corr(), annot=True, cmap='cividis')
plt.show()
```

```python
df['quality'].value_counts().plot(
    kind='bar', 
    figsize=(11, 7), 
    title='Count of Target Classes',
    color='#98ddca'
)
plt.show()
```

```python
df.hist(grid=False, figsize=(15, 11), color='#ffab73')
plt.show()
```

## Split into Features and Target Variable

```python
X = df.loc[:, ('fixed acidity',  'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'wine_type')].values
X
```

```python
y = df.loc[:, 'quality'].values
y
```

## Encoding Categorical Data

### Encode Column wine_type

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [11])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

## Split into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Feature Scaling

columns to scale: fixed acidity, residual sugar, free sulfur dioxide, total sulfur dioxide, alcohol

```python
columns_to_scale = [2, 5, 7, 8, 12]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, columns_to_scale] = sc.fit_transform(X_train[:, columns_to_scale])
X_test[:, columns_to_scale] = sc.transform(X_test[:, columns_to_scale])
```

## Training Model

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=0)
classifier.fit(X_train, y_train)
```

## Evaluate Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(f'accuracy score: {"{:.2f}".format(accuracy_score(y_test, y_pred)*100)} %')
```

### Evaluate with k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
print(f'accuracy score: {"{:.2f}".format(accuracies.mean()*100)} %')
print(f'standard deviation: {"{:.2f}".format(accuracies.std()*100)} %')
```

## Hyperparameter Tuning

```python
classifier.get_params()
```

```python
n_estimators = np.arange(10, 160, 10)
random_state = np.arange(0, 21)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [
    {
        'n_estimators': n_estimators,
        'criterion': ['entropy', 'gini'],
        'random_state': random_state,
    }
]

grid_search = GridSearchCV(
    estimator=classifier,
    param_grid=parameters,
    scoring='accuracy',
    cv=10,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print(f'best accuracy: {"{:.2f}".format(best_accuracy*100)} %')
print(f'best parameters: {best_parameters}')
```

```python
df = pd.DataFrame(grid_search.cv_results_)
df.sort_values('rank_test_score').head(5)
```

```python
y_pred = grid_search.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(f'accuracy score: {"{:.2f}".format(accuracy_score(y_test, y_pred)*100)} %')
```
