## Concatenate All Cars Manufacturer

```python
df = pd.concat([df_audi, df_bmw, df_ford, df_hyundi, df_merc, df_skoda, df_toyota, df_vauxhall, df_vw], ignore_index=True)
df.head()
```

## Split into Features and Target Variable

```python
X = df.loc[:, ('model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize', 'manufacturer')].values
X
```

```python
y = df.loc[:, 'price'].values
y
```

### Encode Column model

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 0] = le.fit_transform(X[:, 0])
X
```

### Encode Column transmission

```python
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
X
```

### Encode Column fuelType

```python
le = LabelEncoder()
X[:, 4] = le.fit_transform(X[:, 4])
X
```

### Encode Column manufacturer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [8])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
df_X = pd.DataFrame(X)
df_X.head()
```

## Split into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Training Model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=11, random_state=1)
regressor.fit(X_train, y_train)
```

## Predict

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

## Evaluate Model

```python
from sklearn.metrics import r2_score, mean_squared_error
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```
